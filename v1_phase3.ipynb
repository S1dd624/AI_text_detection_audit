```python
def find_draft(folder: Path, prefix: str, eid: int):
    matches = list(folder.glob(f"{prefix}_{eid}*"))
    if not matches:
        print(f"[MISS] Draft {prefix}_{eid}")
        return None
    return matches[0]

def find_confirmlog(folder: Path, prefix: str, eid: int):
    matches = list(folder.glob(f"{prefix}_{eid}*"))
    if not matches:
        print(f"[MISS] ConfirmLog {prefix}_{eid}")
        return None
    return matches[0]

def find_annotation_xlsx(folder: Path, eid: int):
    matches = list(folder.glob(f"Annotation_2018argrewrite_{eid}*.xlsx"))
    if not matches:
        print(f"[MISS] Annotation {eid}")
        return None
    # choose most revised version
    return sorted(matches, key=lambda p: len(p.name))[-1]
```

```python
import pandas as pd

SEMANTIC = {
    "Claims/Ideas",
    "Evidence",
    "Warrant/Reasoning/Backing",
    "Rebuttal/Reservation"
}

def transition_is_nonsemantic(annotation_xlsx: Path, confirmlog_txt: Path):
    # revision must be verified
    if confirmlog_txt.read_text().strip() == "":
        return False

    df = pd.read_excel(annotation_xlsx)

    # DEBUG
    print(f"[INFO] Reading {annotation_xlsx.name}")
    print("Columns:", df.columns.tolist())

    # find revision type column (dataset-dependent)
    possible_cols = [c for c in df.columns if "Type" in c or "Revision" in c]

    if not possible_cols:
        print("[WARN] No revision-type column found — conservative reject")
        return False

    for col in possible_cols:
        for val in df[col].dropna().astype(str):
            if val in SEMANTIC:
                return False

    return True
```

```python
def discover_valid_essay_ids():
    ids = []

    for p in D1.iterdir():
        if not p.name.startswith("draft1_2018argrewrite_"):
            continue

        try:
            eid = int(p.stem.split("_")[-1])
        except ValueError:
            continue

        d1 = find_draft(D1, "draft1_2018argrewrite", eid)
        d2 = find_draft(D2, "draft2_2018argrewrite", eid)
        d3 = find_draft(D3, "2018argrewrite", eid)

        a12 = find_annotation_xlsx(A12, eid)
        a23 = find_annotation_xlsx(A23, eid)

        r12 = find_confirmlog(R12, "rev", eid)
        r23 = find_confirmlog(R23, "rev_23", eid)

        if all([d1, d2, d3, a12, a23, r12, r23]):
            ids.append(eid)

    return sorted(ids)

ESSAY_IDS = discover_valid_essay_ids()
len(ESSAY_IDS), ESSAY_IDS
```

```python
from pathlib import Path

def load_text(path: Path):
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read().strip()
```

```python
rows = []

for eid in ESSAY_IDS:
    # --- Drafts (TEXT) ---
    d1 = load_text(D1 / f"draft1_2018argrewrite_{eid}.txt")
    d2 = load_text(D2 / f"draft2_2018argrewrite_{eid}.txt")
    d3 = load_text(D3 / f"2018argrewrite_{eid}.txt")

    # --- Annotations & verification (PATHS, not text) ---
    ann12_path = find_annotation_xlsx(A12, eid)
    ann23_path = find_annotation_xlsx(A23, eid)

    rev12_path = find_confirmlog(R12, "rev", eid)
    rev23_path = find_confirmlog(R23, "rev_23", eid)

    # --- Semantic filtering ---
    valid12 = transition_is_nonsemantic(ann12_path, rev12_path)
    valid23 = transition_is_nonsemantic(ann23_path, rev23_path)

    # --- Detector scores ---
    s1 = sbert_score(d1)
    s2 = sbert_score(d2)
    s3 = sbert_score(d3)

    rows.append({
        "essay_id": eid,
        "AI": AI_SCORE,
        "D1": s1,
        "D2": s2,
        "D3": s3,
        "valid_12": valid12,
        "valid_23": valid23,
        "Δ_AI_D1": s1 - AI_SCORE,
        "Δ_AI_D3": s3 - AI_SCORE,
        "Δ_12": s2 - s1 if valid12 else np.nan,
        "Δ_23": s3 - s2 if valid23 else np.nan,
        "Δ_13": s3 - s1
    })

phase3_df = pd.DataFrame(rows)
phase3_df.head()
```

```python
# Cell: sanity diagnostics
import numpy as np

df = phase3_df.copy()

# Basic stats
print("N essays:", len(df))
print("AI baseline score (fixed):", df['AI'].iloc[0])
print("Human D1 mean:", df['D1'].mean(), "std:", df['D1'].std())
print("Human D2 mean:", df['D2'].mean(), "std:", df['D2'].std())
print("Human D3 mean:", df['D3'].mean(), "std:", df['D3'].std())

# Percent of human drafts scoring higher than AI
pct_d1_higher = (df['D1'] > df['AI']).mean()
pct_d2_higher = (df['D2'] > df['AI']).mean()
pct_d3_higher = (df['D3'] > df['AI']).mean()
print(f"Percent D1 > AI: {pct_d1_higher:.2%}, D2 > AI: {pct_d2_higher:.2%}, D3 > AI: {pct_d3_higher:.2%}")

# Number of essays where D3 > 0.5
print("Count human D3 >= 0.5:", (df['D3'] >= 0.5).sum())
```

```python
# Cell: select only essays where at least one transition is nonsemantic and verified
valid_df = df.copy()
# Use 'valid_12' and 'valid_23' columns (True if safe)
valid_df['any_valid_transition'] = valid_df[['valid_12','valid_23']].any(axis=1)

print("Total essays:", len(df))
print("Essays with at least one verified nonsemantic transition:", valid_df['any_valid_transition'].sum())

# Compute means on this subset
sub = valid_df[valid_df['any_valid_transition']]
print("Subset mean AI_to_D3:", (sub['D3'] - sub['AI']).mean())
print("Subset mean D1_to_D2 (only where valid):", sub.loc[sub['valid_12'],'D2'].mean() - sub.loc[sub['valid_12'],'D1'].mean())
```

```python
# Cell: Wilcoxon signed-rank and paired t-tests: D3 vs AI
from scipy.stats import wilcoxon, ttest_rel

d3_minus_ai = df['D3'] - df['AI']
print("Wilcoxon D3 vs AI p:", wilcoxon(d3_minus_ai).pvalue)
print("Paired t-test D3 vs AI p:", ttest_rel(df['D3'], df['AI']).pvalue)

# For only nonsemantic transitions (D3 if either valid23 or valid12? do D3 vs AI but only where any_valid_transition)
sub = df[df[['valid_12','valid_23']].any(axis=1)]
print("Wilcoxon on verified subset p:", wilcoxon(sub['D3'] - sub['AI']).pvalue)
```

```python
# Cell: effect sizes
df['delta_AI_D3'] = df['D3'] - df['AI']
df_sorted = df.sort_values('delta_AI_D3', ascending=False)
df_sorted[['essay_id','AI','D1','D2','D3','delta_AI_D3']].head(10)
```

```python
# Cell: build labels and compute ROC
from sklearn.metrics import roc_curve, auc
# Build samples: AI (1 sample) + human D3s -> not ideal for ROC (unbalanced), but we can construct an empirical set by repeating the AI or generating a small ensemble.
# Better: compare distributions of AI score vs humans via rank tests (we did). But we can compute empirical sensitivity at thresholds.

human_scores = df['D3'].values
ai_score = df['AI'].iloc[0]

# Fraction of humans above a chosen threshold (e.g., 0.5)
for thr in [0.25,0.5,0.75]:
    print(f"Fraction of humans with score >= {thr}:", (human_scores >= thr).mean())

# Optionally, compute ROC by treating ai_score as a single positive and humans as negatives is unstable; instead compute ranks:
ranks = (human_scores > ai_score).sum()
print("Number of human drafts scoring higher than AI:", ranks)
```

```python
# Cell: compute length and simple features
import re
def char_len(s): return len(s)
def token_count(s): return len(s.split())
def avg_word_len(s): return np.mean([len(w) for w in s.split()]) if s.strip() else 0
def comma_rate(s): return s.count(',') / max(1,len(s.split()))
def question_rate(s): return s.count('?') / max(1,len(s.split()))

for col in ['D1','D2','D3']:
    df[f'{col}_chars'] = df[col].apply(lambda x: char_len(x))
    df[f'{col}_tokens'] = df[col].apply(lambda x: token_count(x))
    df[f'{col}_avgwl'] = df[col].apply(lambda x: avg_word_len(x))

# Correlations with D3 score
print("Spearman corr D3 score vs D3 length:", df['D3'].corr(df['D3_chars'], method='spearman'))
print("Spearman corr D3 score vs D3 avg word len:", df['D3'].corr(df['D3_avgwl'], method='spearman'))
```
